```python
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('country_wise_latest.csv')
df
```

```python
plt.plot(df['Deaths'],df['Recovered'])
```

```python
df.plot(kind='scatter',x='Deaths',y='Recovered')
plt.grid()
plt.plot()
```

```python
df['Country/Region'].value_counts()
```

```python
sns.set(style="darkgrid")
sns.scatterplot(x='Deaths',y='Recovered', hue='WHO Region',data=df)
```

```python
bar_width=35
plt.bar(df['Country/Region'],df['Deaths'],bar_width,label='Dataset1')
plt.bar(df['Country/Region'],df['Recovered'],bar_width-20,label='Dataset2')
plt.xlabel('Country/Region')
plt.ylabel('Deaths')
plt.legend()
plt.show()
```

```python
sns.relplot(x='Deaths',y='Recovered',hue='WHO Region',data=df)
```

```python
sns.catplot(x='Deaths',y='Recovered',hue='WHO Region',kind='bar',data=df)
```

```python
sns.catplot(x='Deaths',y='Recovered',hue='WHO Region',kind='violin',data=df)
```

```python
sns.catplot(x='Deaths',y='Recovered',hue='WHO Region',kind='box',data=df)
```

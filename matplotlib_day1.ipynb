```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.linspace(0,100,10)
y=x*x
print("x = ",x)
print("y = ",y)
plt.plot(x,y)
```

```python
#Add labels and title
x=['A','B','C']
y=[50,70,90]
print("X = ",x)
print("Y = ",y)
plt.title("Line chart Demo")
plt.xlabel("X-axis-Category")
plt.ylabel("Y-axis-Numbers")
plt.plot(x,y,color='green',marker="*",linestyle="dashed",mec="r",mfc="b",linewidth=3,markersize=20)
#marker_edge_color(mec)
#marker_face_color(mfc)
```

```python
x=np.linspace(0,10,100)
y=np.sin(x)
y2=np.tan(x)
plt.plot(x,y)
plt.plot(x,y2)
```

```python
#generate somedata
x=np.array(['A','B','C','D'])
y=np.array([3,8,1,10])
plt.plot(x,y)
```

```python
#For categorial values ->bar charts are suitable
#create the plot
plt.bar(x,y)
```

```python
plt.barh(x,y) #barh->horizontal bar chart
```

```python
categories=np.array(['A','B','C','D'])
value1=np.array([3,8,1,10])
value2=np.array([6,2,7,5])
value3=np.array([10,4,8,1])
#set the width of the bars
bar_width=0.35
#create the plot
plt.bar(categories,value1,bar_width,label='Dataset 1')
plt.bar(categories,value2,bar_width-0.20,label='Dataset 2')
plt.bar(categories,value3,bar_width-0.3,label='Dataset 3')
#Add levels and title
plt.xlabel('Category')
plt.ylabel('Value')
plt.title('Grouped Bar Chart Example')
#Add legend
plt.legend()
#Dislpay the plot
plt.show()
```

```python
#pi chart ->for composition
categories=np.array(['A','B','C','D'])
values=np.array([3,8,1,10])
# create the plot
plt.pie(values, labels=categories)
#Add title
plt.title('Simple pie chart Example')
#Display the plot
plt.show
```

```python
#Explode -> for highlight the slice
#define which slice to explode
explode=(0,0,0.1,0)
#create the plot
plt.pie(values,labels=categories,explode=explode)
#Add title
plt.title('Exploded Pie chart Example')
#display the plot
plt.show()
```

```python
Items=np.array(['Electronics','Clothing','Grocery','House-Hold'])
Sell=np.array([5000,6000,4000,3500])
plt.title("Items and Sell")
plt.xlabel("Items")
plt.ylabel("Sell")
bar_width=0.35
plt.bar(Items,Sell,bar_width)
plt.show()
```

```python
Items=np.array(['Electronics','Clothing','Grocery','House-Hold'])
Sell=np.array([5000,6000,4000,3500])
plt.title("Items and Sell")
plt.xlabel("Items")
plt.ylabel("Sell")
bar_width=0.35
grocery=(0,0,.2,0)
plt.pie(Sell,labels=Items ,explode=grocery,autopct="%1.1f%%")
#autopct ->auto percentage
plt.show()
```

```python
#Donut chart example
categories=np.array(['A','B','C','D'])
values=np.array([3,8,1,10])
fig, ax=plt.subplots()
#draw the outer ring
ax.pie(values,labels=categories,radius=1 ,startangle=90,counterclock=False,)
#draw the inner ring
ax.pie([1],radius=0.6,colors='w')
#Add title
ax.set_title("Donut Chart Example")
plt.show()
```
